```python
# all imports 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# Reading data
stopWords = set(stopwords.words("english"))
text = open('bachelor.txt').read()
words = word_tokenize(text)
```

```python
# appling preprocessing techniques and calculating term frequency
freqTable = dict()
for word in words:
    word = word.lower()
    if word in stopWords:
        continue
    if word in freqTable:
        freqTable[word] += 1
    else:
        freqTable[word] = 1
```

```python
# breaking into sentences
sentences = sent_tokenize(text)
sentenceValue = dict()
```

```python
# dictionary keys = sentences, value = frequency of words
for sentence in sentences:
    for wordValue in freqTable:
        if wordValue in sentence.lower():
            if sentence[:12] in sentenceValue:
                sentenceValue[sentence[:12]] += freqTable[wordValue]
            else:
                sentenceValue[sentence[:12]] = freqTable[wordValue]
```

```python
sumValues = 0
for sentence in sentenceValue:
    sumValues += sentenceValue[sentence]

# Average value of a sentence from original text
average = int(sumValues/ len(sentenceValue))
```

```python
summary = ''
for sentence in sentences:
        if sentence[:12] in sentenceValue and sentenceValue[sentence[:12]] > (1.5 * average):
            summary +=  " " + sentence
print(summary)
```
